```python
import glob
import json
import os

from aizynthfinder.reactiontree import ReactionTree

OUTPUT_DIR = "/data/rlawlsgurjh/work/gitlab/CUE/cue_vs/results/6S89/CUE/Trial_1/Proxy/PI/aizynthfinder/"
IMAGES_DIR = os.path.join(OUTPUT_DIR, "tree_images")

os.makedirs(IMAGES_DIR, exist_ok=True)
```

```python
json_files = sorted(glob.glob(os.path.join(OUTPUT_DIR, "*.json")))
print(f"Found {len(json_files)} json file(s) in {OUTPUT_DIR}")
```

```python
failed = []
for json_path in json_files:
    mol_id = os.path.splitext(os.path.basename(json_path))[0]

    with open(json_path) as f:
        trees = json.load(f)

    if not trees:
        print(f"[{mol_id}] no routes found, skipping")
        failed.append(mol_id)
        continue

    best_route = max(trees, key=lambda route: route["scores"]["state score"])
    image = ReactionTree.from_dict(best_route).to_image()

    image_path = os.path.join(IMAGES_DIR, f"{mol_id}.png")
    image.save(image_path)
    print(f"[{mol_id}] saved -> {image_path}")

print(f"Done. {len(json_files) - len(failed)} saved, {len(failed)} failed: {failed}")
```
